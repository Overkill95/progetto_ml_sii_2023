## Progetto ML 2023/2024 - lettura dati

```python
import pandas as pd
import os
import numpy as np
import math
google_reviews_data = pd.read_csv(os.path.join('.', "google_review_ratings.csv"))
```

```python
google_reviews_data
```

```python
split_size = np.shape(google_reviews_data)[0]
split_limit = split_size - math.floor(split_size * 0.25)
split_limit
```

```python
google_reviews_data_training = google_reviews_data[google_reviews_data.index < split_limit]
google_reviews_data_test = google_reviews_data[google_reviews_data.index >= split_limit]
```

```python
google_reviews_data = google_reviews_data.dropna(axis='columns')
google_reviews_data_training = google_reviews_data_training.dropna(axis='columns')
google_reviews_data_test = google_reviews_data_test.dropna(axis='columns')
```
